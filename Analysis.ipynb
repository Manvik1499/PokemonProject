```python
import pandas as pd
import matplotlib.pyplot as plt

pokemonBasicData = pd.read_csv("./pokemonBasicData")
bmiList = list(pokemonBasicData.BMI)
pokemonId = [i for i in range(len(bmiList))]

plt.plot(pokemonId,bmiList)
plt.show()
```

```python
# Creating plot
red_circle = dict(markerfacecolor='red', marker='o')
plt.boxplot(bmiList,vert=False, flierprops=red_circle)
 
# show plot
plt.show()
```
